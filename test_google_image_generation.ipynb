# Testing Google Image Generation Models

This notebook tests all available Google image generation models using the `get_image_generation_completion` function from `utils.py`. Each model will generate an image based on a test prompt, display it inline in the notebook, and save it to disk for use with `get_vision_completion`.

## Requirements
- Google API key configured in `.env` file as `GOOGLE_API_KEY` or `GEMINI_API_KEY`
- Required Python packages: `google-genai`, `google.generativeai`, `IPython`, `PIL`
- Internet connection for API calls

1. `gemini-2.5-flash-image-preview` - Gemini with image generation
2. `imagen-3.0-generate-002` - Google's Imagen 3.0
3. `imagen-4.0-generate-001` - Google's Imagen 4.0

````python
# Import Required Libraries
import sys
import os

# Add the current directory to Python path to import utils
sys.path.append(os.getcwd())

# Import functions from utils.py
from utils import (
    load_environment,
    setup_llm_client,
    get_image_generation_completion,
    get_vision_completion
)

# Import IPython display functions
from IPython.display import display, Image, Markdown, HTML
import time
````

````python
# Load Environment
# This will load API keys from the .env file
load_environment()
print("‚úÖ Environment loaded successfully!")
````

````python
# Define Test Prompt
# This prompt will be used for all image generation tests
test_prompt = "A beautiful sunset over mountains with a lake, vibrant colors, photorealistic style, high detail"

print(f"Test prompt: {test_prompt}")
print("This prompt will be used to test all Google image generation models.")
````

## Test 1: Gemini 2.5 Flash Image Preview

Testing the `gemini-2.5-flash-image-preview` model for image generation capabilities.

````python
# Test Gemini 2.5 Flash Image Preview
model_name = "gemini-2.5-flash-image-preview"

print(f"üîÑ Setting up client for {model_name}...")
client, model, provider = setup_llm_client(model_name)

if client:
    print(f"‚úÖ Client setup successful for {provider}")
    print(f"üîÑ Generating image with {model_name}...")

    file_path, image_url = get_image_generation_completion(test_prompt, client, model, provider)

    if file_path and image_url:
        print(f"‚úÖ Image generated successfully!")
        print(f"üìÅ Saved to: {file_path}")
        print(f"üñºÔ∏è  Image URL ready for get_vision_completion: {file_path}")

        # Display the image in the notebook
        display(Image(url=image_url, width=400))
    else:
        print("‚ùå Image generation failed")
else:
    print(f"‚ùå Failed to setup client for {model_name}")
````

## Vision Analysis Demo

Now let's demonstrate how to use `get_vision_completion` with one of the generated images. We'll analyze the first successfully generated image.

````python
# Vision Analysis Demo
# This demonstrates how to use get_vision_completion with a generated image

# First, let's find a generated image to analyze
import glob
import os

generated_images = glob.glob("artifacts/screens/image_*.png")
if generated_images:
    # Use the most recently generated image
    test_image_path = max(generated_images, key=os.path.getctime)

    print(f"üîç Analyzing image: {test_image_path}")

    # Set up a vision-capable model for analysis
    vision_client, vision_model, vision_provider = setup_llm_client("gemini-2.5-flash")

    if vision_client:
        # Ask about the generated image
        vision_prompt = "Describe this image in detail. What do you see? What is the mood and style?"

        print(f"ü§ñ Analyzing image with {vision_model}...")
        analysis = get_vision_completion(vision_prompt, test_image_path, vision_client, vision_model, vision_provider)

        if analysis:
            print("‚úÖ Vision analysis completed!")
            print("\nüìù Analysis:")
            print(analysis)

            # Display the analyzed image
            display(Image(filename=test_image_path, width=400))
        else:
            print("‚ùå Vision analysis failed")
    else:
        print("‚ùå Could not setup vision client for analysis")
else:
    print("‚ùå No generated images found for analysis")
````

All generated images are saved in the `artifacts/screens/` directory with timestamps.
You can use any of these paths directly with `get_vision_completion()`:

```python
# Example usage:
vision_client, vision_model, vision_provider = setup_llm_client("gemini-2.5-flash")
analysis = get_vision_completion(
    "Describe this image",
    "artifacts/screens/image_[timestamp].png",
    vision_client, vision_model, vision_provider
)
```
